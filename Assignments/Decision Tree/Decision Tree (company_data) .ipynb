# Aim:
 A cloth manufacturing company is interested to know about the segment or attributes causes high sale. Approach - A decision tree can be built with target variable Sale (we will first convert it in categorical variable) & all other variable will be independent in the analysis.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder
%matplotlib inline
```

```python
df=pd.read_csv('Company_data.csv')
df.head()
```

# EDA

```python
df.describe()
```

```python
df['sales'] = np.where((df['Sales'] < 8), 'Low', 'High')
df
```

```python
df1 = df.iloc[:,1:]
df1
```

```python
oe = OrdinalEncoder()
le = LabelEncoder()

df1.iloc[:,[5,8,9]] = oe.fit_transform(df1.iloc[:,[5,8,9]])
df1['sales'] = le.fit_transform(df['sales'])
df1
```

# Splitting the Data

```python
x = df1.iloc[:,1:]
y = df1['sales']
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=1)
```

```python
dt = DecisionTreeClassifier(criterion='entropy')
```

```python
dt.fit(xtrain, ytrain)

ypred = dt.predict(xtest)
```

```python
from sklearn.metrics import classification_report

print(classification_report(ytest, ypred))
```

```python
print('Training accuracy: ', dt.score(xtrain,ytrain))
print('Testing Accuracy: ', dt.score(xtest,ytest))
```

# Creating tree diagram

```python
tree.plot_tree(dt)
plt.show()
```

```python
tree.plot_tree(dt, feature_names=x.columns, class_names=df['sales'].unique(), filled=True)
plt.show()
```
