# Aim:
Forecast the CocaCola prices data set. Prepare a document for each model explaining how many dummy variables you have created and RMSE value for each model. Finally which model you will use for Forecasting.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as smf
from sklearn.metrics import mean_squared_error
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
df= pd.read_excel('CocaCola_Sales_Rawdata.xlsx')
df
```

## EDA

```python
df.info()
```

```python
df.describe(include='all')
```

```python
# Extracting quaters and years values
df['Quarters'] = 0
df['Year'] = 0
for i in range(42):
    p = df["Quarter"][i]
    df['Quarters'][i]= p[0:2]
    df['Year'][i]= p[3:5]
```

```python
dum = pd.get_dummies(df['Quarters'])
```

```python
df = pd.concat([df, dum], axis=1)
df
```

```python
# Data for models
df.drop(columns='Quarter', inplace=True)
df['t'] = np.arange(1,43)
df['t_square'] = np.square(df['t'])
df['log_sales'] = np.log(df['Sales'])
df
```

## Splitting Data

```python
x = df.head(32)
y = df.tail(10)
```

## Model based Forecasting Methods

```python
df.head(2)
```

## Linear model

```python
linear = smf.ols('Sales~t', data=x).fit()
pred_linear = pd.Series(linear.predict((y[['t']])))
rmse_linear = np.sqrt(mean_squared_error(y['Sales'], pred_linear))
rmse_linear
```

## Exponential model

```python
exp = smf.ols('log_sales~t', data=x).fit()
pred_exp = pd.Series(exp.predict(y[['t']]))
rmse_exp = np.sqrt(mean_squared_error(y['Sales'], pred_exp))
rmse_exp
```

## Quadratic model

```python
quad = smf.ols('Sales~t+t_square',data=x).fit()
pred_quad = pd.Series(quad.predict(y[['t', 't_square']]))
rmse_quad = np.sqrt(mean_squared_error(y['Sales'], pred_quad))
rmse_quad
```

## Additive seasonality model

```python
add_sea = smf.ols('Sales~Q1+Q2+Q3+Q4',data=x).fit()
pred_add_sea = pd.Series(add_sea.predict(y[['Q1', 'Q2', 'Q3', 'Q4']]))
rmse_add_sea = np.sqrt(mean_squared_error(y['Sales'], pred_add_sea))
rmse_add_sea
```

## Additive seasonality quadratic model

```python
add_sea_quad = smf.ols('Sales~t+t_square+Q1+Q2+Q3+Q4',data=x).fit()
pred_add_sea_quad = pd.Series(add_sea_quad.predict(y[['Q1', 'Q2', 'Q3', 'Q4', 't', 't_square']]))
rmse_add_sea_quad = np.sqrt(mean_squared_error(y['Sales'], pred_add_sea_quad))
rmse_add_sea_quad
```

## Multiplicative seasonality model

```python
mul_sea = smf.ols('log_sales~Q1+Q2+Q3+Q4', data=x).fit()
pred_mul_sea = pd.Series(mul_sea.predict(y[['Q1', 'Q2', 'Q3', 'Q4']]))
rmse_mul_sea = np.sqrt(mean_squared_error(y['Sales'], pred_mul_sea))
rmse_mul_sea
```

## Multiplicative additive seasonality model

```python
mul_add_sea = smf.ols('log_sales~t+Q1+Q2+Q3+Q4', data=x).fit()
pred_mul_add_sea = pd.Series(mul_add_sea.predict(y[['Q1', 'Q2', 'Q3', 'Q4', 't']]))
rmse_mul_add_sea = np.sqrt(mean_squared_error(y['Sales'], pred_mul_add_sea))
rmse_mul_add_sea
```

## Results

```python
data = {
    'Model' : ['Linear model', 'Exponential model', 'Quadratic model', 'Additive seasonality model',
               'Additive seasonality quadratic model', 'Multiplicative seasonality model', 'Multiplicative additive seasonality model'],
    'RMSE Values' : [rmse_linear, rmse_exp, rmse_quad, rmse_add_sea, rmse_add_sea_quad, rmse_mul_sea, rmse_mul_add_sea]
}
result = pd.DataFrame(data)
result
```

## Building final model with least RMSE value ie. Additive seasonality quadratic model

```python
final_model = smf.ols('Sales~t+t_square+Q1+Q2+Q3+Q4',data=df).fit()
pred_final_model = pd.Series(final_model.predict(df[['Q1', 'Q2', 'Q3', 'Q4', 't', 't_square']]))
rmse_final_model = np.sqrt(mean_squared_error(df['Sales'], pred_final_model))
rmse_final_model
```
